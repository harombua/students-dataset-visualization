### 2.1) Domain analysis: Understanding the features (columns)

```python
# Understanding 
student_data.columns
```

1. school:-student's school (binary: 'GP' - Gabriel Pereira or 'MS' - Mousinho da Silveira)
2. sex:-student's sex (binary: 'F' - female or 'M' - male)
3. age:-student's age (numeric: from 15 to 22)
4. address:-student's home address type (binary: 'U' - urban or 'R' - rural)
5. famsize:-family size (binary: 'LE3' - less or equal to 3 or 'GT3' - greater than 3)
6. Pstatus:-parent's cohabitation status (binary: 'T' - living together or 'A' - apart)
7. Medu:-mother's education (numeric: 0 - none, 1 - primary education (4th grade), 2 â€“ 5th to 9th grade, 3 â€“ secondary
8. Fedu:-father's education (numeric: 0 - none, 1 - primary education (4th grade), 2 â€“ 5th to 9th grade, 3 â€“ secondary
9. Mjob:-mother's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home' 
10. Fjob:-father's job (nominal: 'teacher', 'health' care related, civil 'services' (e.g. administrative or police), 'at_home'
11. reason:-reason to choose this school (nominal: close to 'home', school 'reputation', 'course' preference or 'other')
12. guardian:-student's guardian (nominal: 'mother', 'father' or 'other')
13. traveltime:-home to school travel time (numeric: 1 - <15 min., 2 - 15 to 30 min., 3 - 30 min. to 1 hour, or 4 - >1 hour)
14. studytime:-weekly study time (numeric: 1 - <2 hours, 2 - 2 to 5 hours, 3 - 5 to 10 hours, or 4 - >10 hours)
15. failures:- number of past class failures (numeric: n if 1<=n<3, else 4)
16. schoolsup:-extra educational support (binary: yes or no)
17. famup:-family educational support (binary: yes or no)
18. paid:-extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
19. activities:-extra-curricular activities (binary: yes or no)
20. nursery:-attended nursery school (binary: yes or no)
21. higher:-wants to take higher education (binary: yes or no)
22. internet:-Internet access at home (binary: yes or no)
23. romantic:-with a romantic relationship (binary: yes or no)
24. famrel:-quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
25. freetime:-free time after school (numeric: from 1 - very low to 5 - very high)
26. goout:-going out with friends (numeric: from 1 - very low to 5 - very high)
27. Dalc:-workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
28. Walc:-weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
29. health:-current health status (numeric: from 1 - very bad to 5 - very good)
30. absences:-number of school absences (numeric: from 0 to 93)
31. G1:-first period grade (numeric: from 0 to 20)
32. G2:-second period grade (numeric: from 0 to 20)
33. G3:-final grade (numeric: from 0 to 20, output target)

```python
# percentage of those supported in each school
sup_GPstudent=student_data[(student_data.school=='GP') & (student_data.schoolsup=='yes')].school.count()
all_GPstudent =student_data[(student_data.school=='GP')].school.count()
pct_GPsupported=sup_GPstudent/all_GPstudent
print(f'The percentage of the supported students in GP School is: {pct_GPsupported:.1%}.')
sup_MSstudent=student_data[(student_data.school=='MS') & (student_data.schoolsup=='yes')].school.count()
all_MSstudent =student_data[(student_data.school=='MS')].school.count()
pct_MSsupported=sup_MSstudent/all_MSstudent
print(f'The percentage of the supported students in MS School is: {pct_MSsupported:.1%}.')
```

* Considering the address and traveltime, it appears that GP is near an urban area, while MS is near a rural area since we have more students from urban area in GP who seem to travel in a short distance to arrive at school, while we have more students from rural area who also take short time to arrive at school. This helps us to confirm why GP is more populated than MS-it is near an urban area.
* Considering the extra educational support factor, we realise that in GP we have more students having the extra support as compared to those in MS, this may influence more students to attend GP as compared to MS. * Considering the age factor as showm in the graph above, we realise that GP school has older ages, meaning there possibly could be more higher classes as compared to MS school which could imply more population in school GP than in MS.

### Question 3: Performance analysis.

```python
student_data['performance']= (student_data.G1+student_data.G2+student_data.G3)/3
student_data.sort_values('performance',ascending=False)
```

```python
student_data.performance.mean()
```

```python
plt.figure(figsize=(12,6))
sn.scatterplot(data=student_data,y='age',x='performance',s=50,hue='school');
```

```python
# Calculating the mean score to determine those above mean score school-wise and the calculate the percentage.
student_data[student_data.performance>=11.62].school.value_counts()*100/student_data.groupby('school').performance.count()
```

### Insights:
From the above scatter plot, we can observe that students at MS performed poorly (Only 32% were above the mean score) as compared to those at the GP school where 60.3% score above the mean score.

This could also be another reason as to why the school MS has a less population compared to the school GP.

```python
sn.barplot(data=student_data,x='school',y='performance',hue='sex',palette=['green','orange']);
```

```python
# Comparing gender in GP school to see their performance
GPgirls_mean=student_data[(student_data.school=='GP') & (student_data.sex=='F')].performance.sum()/student_data[(student_data.school=='GP') & (student_data.sex=='F')].performance.count()
GPboys_mean=student_data[(student_data.school=='GP') & (student_data.sex=='M')].performance.sum()/student_data[(student_data.school=='GP') & (student_data.sex=='M')].performance.count()
print('GP Girls mean:',GPgirls_mean,'\n','GP Boys Mean:',GPboys_mean)
```

```python
# Comparing gender in MS school to see their performance
MSgirls_mean=student_data[(student_data.school=='MS') & (student_data.sex=='F')].performance.sum()/student_data[(student_data.school=='MS') & (student_data.sex=='F')].performance.count()
MSboys_mean=student_data[(student_data.school=='MS') & (student_data.sex=='M')].performance.sum()/student_data[(student_data.school=='MS') & (student_data.sex=='M')].performance.count()
print('MS Girls mean:',MSgirls_mean,'\n','MS Boys Mean:',MSboys_mean)
```

### Insights:
As shown in the barplot above, in each school, girls seem to perform better than boys, as shown the GP girls have a mean score of 12.6, while boys have a mean of 11.8 of which both are above the overal mean. In MS school, both means are less than the overal mean. Girls lead with a meanscore of 10.8, while boys have a mean of 9.9.
